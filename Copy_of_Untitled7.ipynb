### Handling Missing data

```python
# Check for missing values
df.isnull().sum()
```

```python
# filling the categorical columns with mode
df['Engine Fuel Type'].fillna(df['Engine Fuel Type'].mode()[0], inplace=True)
df['Number of Doors'].fillna(df['Number of Doors'].mode()[0], inplace=True)
print(df)
```

```python
# filling the numerical columns with appropriate statistics
df['Engine HP'].fillna(df['Engine HP'].median(), inplace=True)
df['Engine Cylinders'].fillna(df['Engine Cylinders'].mode()[0], inplace=True)
print(df)
```

### Data type convention

```python
df['Year'] = df['Year'].astype(int)  #'Year' is an integer
df['Engine HP'] = df['Engine HP'].astype(float)  #'Engine HP' is a float
df['Engine Cylinders'] = df['Engine Cylinders'].astype(float)  #'Engine Cylinders' is a float
df['Number of Doors'] = df['Number of Doors'].astype(float)  #'Number of Doors' is a float
df['highway MPG'] = df['highway MPG'].astype(float)  #'highway MPG' is a float
df['city mpg'] = df['city mpg'].astype(float)  #'city mp
```

```python
print(df.dtypes) # verifying the correctness of the dataset
```

### Filtering Data

```python
df_filtered = df[df['Year'] >= 1995]   # filtering the dataset to show only cars from 1995 and later
print(df_filtered)
```

### Visualization

### Histogramm

```python
# plotting histogram for city mpg
plt.figure(figsize=(10, 6))
plt.hist(df['city mpg'], bins=20, color='green', edgecolor='black')
#adding labels and title
plt.xlabel('City MPG')
plt.ylabel('Frequency')
plt.title('Distribution of City MPG')
plt.grid(axis='y', alpha=0.7)
# show the histogram
plt.show()
```

### Bar Chart

```python
# calculating average msrp for each vehicle size category
avg_msrp_per_size = df.groupby('Vehicle Size')['MSRP'].mean().sort_values()
#plotting the bar chart
plt.figure(figsize=(10, 6))
avg_msrp_per_size.plot(kind='bar', color='green', edgecolor='black')
#adding labels and title
plt.xlabel('Vehicle Size')
plt.ylabel('Avg MSRP ($)')
plt.title('Average MSRP by Vehicle Size')
plt.xticks(rotation=45)
plt.show()
```

### Scatterd Plot

```python
#plottiing scatter plot for Engine HP vs MSRP
plt.figure(figsize=(10, 6))
plt.scatter(df['Engine HP'], df['MSRP'], color='green', alpha=0.5)
#adding labels and title
plt.xlabel('Engine HP')
plt.ylabel('MSRP ($)')
plt.title('Relationship between Engine HP and MSRP')
plt.grid(True, alpha=0.7)
plt.show()
```

### Box plot

```python
#setting figure size
plt.figure(figsize=(10, 6))
#creating a box plot
sns.boxplot(x='Driven_Wheels', y='MSRP', data=df)
#adding labels and title
plt.xlabel('Driven_Wheels')
plt.ylabel('MSRP ($)')
plt.title('MSRP Distribution by Driven_Wheels')
plt.yscale('log') # log scale is to handle large MSRP range
plt.xticks(rotation=45)
plt.show()
```

## line Plot

```python
# group by transmission type and calculating the avg city and highway mpg
mpg_trends = df.groupby('Transmission Type')[['city mpg', 'highway MPG']].mean()
#plotting the line plot
plt.figure(figsize=(10, 6))
plt.plot(mpg_trends.index, mpg_trends['city mpg'], marker='o', label='City MPG', color='green')
plt.plot(mpg_trends.index, mpg_trends['highway MPG'], marker='s', label='Highway MPG', color='red')
# adding labels and title
plt.xlabel('Transmission Type')
plt.ylabel('Average MPG')
plt.show()
```

### Corrolation metrix

```python
#selecting relevant columns
columns = ['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']
# computing the correlation metrix
correlation_matrix = df[columns].corr()
#printing the correlation matrix
print('Correlation Matrix:\n', correlation_matrix)
# plotting a heatmap for better visualization
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
#adding title
plt.title('Correlation Matrix of Selected Variables')
plt.show()
```

The analysis shows clear patterns in car pricing based on size. Bigger vehicles usually cost more, but there are some exceptions. When looking at Engine HP (horsepower) vs. MSRP (manufacturers suggested retail price), there is a clear trendcars with more horsepower are generally more expensive. However, a few outliers suggest that luxury or high-performance features also play a role in pricing.

A boxplot comparing MSRP across different drivetrains (how the wheels are powered) shows a big difference in prices. All-wheel-drive (AWD) vehicles usually have a higher median price compared to front-wheel-drive (FWD) and rear-wheel-drive (RWD) cars.

Fuel efficiency also varies based on transmission type. Cars with automatic transmissions tend to have better miles per gallon (MPG) in both city and highway driving compared to manual ones.

A correlation analysis confirms some strong relationships. Engine HP and MSRP are closely linked, meaning cars with more power usually cost more. On the other hand, city and highway MPG have a negative correlation with MSRP. This means that more expensive cars often have lower fuel efficiency.

In summary, larger vehicles and those with more horsepower tend to have higher prices. AWD vehicles generally cost more than FWD and RWD models. Automatic transmissions offer better fuel efficiency than manual ones. Finally, cars that cost more are often less fuel-efficient.
